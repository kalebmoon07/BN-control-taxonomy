```python
from bntaxonomy.instance_gen import InstanceGen
from itertools import pairwise
import os,json
```

```python
inst_gen = InstanceGen("ce_long_attr", 3)
for v1, v2 in pairwise(["000", "001", "101", "111", "011", "010", "110", "100"]):
    inst_gen.add_edge(v1, v2)
inst_gen.add_edge("100", "000")
print(inst_gen.to_bnet())
fpath = "instances/counterexamples/ce_long_attr"
inst_gen.save(fpath, target={"x1": 0, "x2": 0, "x3": 0})
inst_gen.show_stg()
```

```python
inst_gen = InstanceGen("ce_long_attr",2)
for edge in [("00","01"),("00","10"), ("10","11")]:
    inst_gen.add_edge(*edge)
print(inst_gen.to_bnet())
fpath = "instances/counterexamples/ce_yes_PTC_no_TTC"
inst_gen.save(fpath,target={"x1": 1, "x2": 1})
inst_gen.show_stg()
```
